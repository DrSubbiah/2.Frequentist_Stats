Let $X_1,X_2,...,X_n$ be iid random sample from normal distribution with mean $\theta$ and variance ${\sigma}^2$ 

### <font color = darkblue> Case 1 : Single mean, where ${\sigma}^2$ is known

#### $\theta \leq \theta_0$ Vs $\theta > \theta_0$

Acceptance region is 
$\bar{x} - \theta_0 \leq z_\alpha \frac{\sigma}  {\sqrt n}$

Corrosponding interval is 
$[\bar{x}-z_\alpha \frac{\sigma}{\sqrt n},\infty)$

#### $\theta \geq \theta_0$ Vs  $\theta < \theta_0$

Acceptance region is 
$\bar{x} - \theta_0 > z_{1-\alpha} \frac{\sigma}  {\sqrt n}$ or

$\bar{x} - \theta_0 > -z_\alpha \frac{\sigma}  {\sqrt n}$

$\Rightarrow \bar{x} + z_\alpha \frac{\sigma}  {\sqrt n} > \theta_0$

Interval is 
$(-\infty,\bar{x} + z_\alpha \frac{\sigma}  {\sqrt n})$

#### $\theta = \theta_0$ Vs $\theta \neq \theta_0$ 

Acceptance Region 
$|\bar{x} - \theta_0| \leq z_{\alpha/2} \frac{\sigma}  {\sqrt n}$

$\Rightarrow - z_{\alpha / 2} \frac{\sigma}{\sqrt n } \leq \bar{x} - \theta_0 \leq z_{\alpha /2}\frac{\sigma}{\sqrt n }$

$\Rightarrow$ Interval is 
$(\bar{x} - z_{\alpha / 2} \frac{\sigma}{\sqrt n } , \bar{x} + z_{\alpha / 2} \frac{\sigma}{\sqrt n })$
 
**<font color = red>If ${\sigma^2}$ is unknown, Replace $z$ by $t_{n-1}$ and $\sigma^2$ by the sample variances $S^2$ **
 
 
### <font color = darkblue> Case 2 : Independent Two Sample ($\sigma_1^2$,$\sigma_2^2$ Known)
 
#### $\theta_1 - \theta_2 \leq \delta$ Vs $\theta_1 - \theta_2 > \delta$

Acceptance Region 
$\bar{x} - \bar{y} - \delta \leq z_\alpha  \sqrt {\frac{\sigma_1^2}{m} +   \frac{\sigma_2^2}{n}}$

Confidence interval is 
$[\bar{x}-\bar{y} -z_\alpha \sqrt {\frac{\sigma_1^2}{m} + \frac{\sigma_2^2}{n}}~, ~~\infty)$
 
#### $\theta_1 - \theta_2 \geq \delta$ Vs $\theta_1 - \theta_2 < \delta$

Acceptance Region 
$\bar{x} - \bar{y} - \delta \geq  z_{1-\alpha}  \sqrt {\frac{\sigma_1^2}{m} +   \frac{\sigma_2^2}{n}}$

$= - z_\alpha  \sqrt {\frac{\sigma_1^2}{m} +   \frac{\sigma_2^2}{n}}$

Confidence interval is 
$( -\infty~ ,~~ \bar{x}-\bar{y} + z_\alpha \sqrt {\frac{\sigma_1^2}{m} + \frac{\sigma_2^2}{n}}]$
 
#### $\theta_1 - \theta_2 = \delta$ Vs $\theta_1 - \theta_2 \neq  \delta$

Confidence interval is 
$(\bar{x}-\bar{y} -  z_{\alpha/2} \sqrt {\frac{\sigma_1^2}{m} + \frac{\sigma_2^2}{n}},~~\bar{x}-\bar{y} + z_{\alpha/2} \sqrt {\frac{\sigma_1^2}{m} + \frac{\sigma_2^2}{n}})$

**<font color = red>if $\sigma_1^2 , \sigma_2^2$ are unknown Replace $z$ by $t_{m+n-2}$ and 
$SE = S_p \sqrt{\frac{1}{m}+\frac{1}{n}}, ~~ S_p^2 = \frac{(m-1)S_1^2 + (n-1)S_2^2}{m+n-2}$**
 
### <font color = darkblue> Case 3 : Bivariate Samples

use Case 1 to have intervals with $W_i = x_i - y_i$

$S_W =\frac{\sum (W_i - W)^2}{n-1}$

$\bar{W}=\bar{x}-\bar{y}$ 

$[ \bar{W} - z_\alpha \frac{S_W}{\sqrt n} , \infty)$

$(-\infty , \bar{W} + z_\alpha \frac{S_W}{\sqrt n}]$

$(\bar{W} - z_\alpha \frac{S_W}{\sqrt n} , \bar{W} + z_\alpha \frac{S_W}{\sqrt n})$


### <font color = darkblue> Case 4 : Single Variance in Normal$(\mu, \sigma^2)$

**<font color = red>1. $\mu$ is known **

$X_1,X_2,...,X_n  \sim Normal~(\mu,\theta) ~~~~~\theta = \sigma^2$ and $\mu$ is Known

#### $\theta \leq \theta_0$ Vs  $\theta > \theta_0$

Acceptance Region is 
$\frac {\sum(x_i-\mu)^2}{\theta_0} < \chi^2_{\alpha,n}$

$\Rightarrow$ Interval is 
$[\frac{\sum(x_i-\mu)^2 }{\chi^2_{\alpha,n}}, \infty )$  for $(\theta=\sigma^2)$

Or,

$[\sqrt{\frac{\sum(x_i-\mu)^2}{\chi^2_{\alpha,n}}},\infty)$ for $\theta = \sigma$

#### $\theta \geq \theta_0$ Vs  $\theta < \theta_0$

Acceptance Region 
$\frac{\sum(x_i-\mu)^2}{\theta_0} > \chi^2_{1-\alpha,n}$

Interval is $\Big(0, \frac{\sum(x_i-\mu)^2}{\chi^2_{(1-\alpha),n}}\Big]$

#### $\theta =  \theta_0 ~~ Vs~~  \theta \neq \theta_0$

Interval is
$\Big(\frac{\sum(x_i-\mu)^2}{\chi^2_{n,\alpha/2}},\frac{\sum(x_i-\mu)^2}{\chi^2_{(1-\alpha /2 ),n}}\Big)$

**<font color = red>2. $\mu$ is unknown **

Replace $\mu$ by $\bar{x}$ and the degrees of freedom is $n - 1$


### <font color = darkblue> Case 5 : Equality of variances in Normal$(\mu_1, \sigma_1^2)$ and Normal $(\mu_2, \sigma_2^2)$

**<font color = red>1. $\mu_1~\&~\mu_2$ are known **

$X_1,...,X_n \sim Normal~(\mu_1,\sigma_1^2)$ and $Y_1,...,Y_n \sim Normal~(\mu_2,\sigma_2^2)$

**TEST 1**

$\Big[\frac{1}{F_{\alpha,m,n}} \frac{\sum(x_i-\mu_1)^2}{\sum(y_j-\mu_2)^2} \frac{n}{m},\infty)$

**TEST 2**

$(0, \frac{1}{F_{1-\alpha,m,n}} \frac{\sum(x_i-\mu_1)^2}{\sum(y_j-\mu_2)^2} \frac{n}{m}\Big]$

**TEST 3**

$\Big(\frac{1}{F_{\alpha/2,m-1,n-1}} \frac{\sum(x_i-\mu_1)^2}{\sum(y_j-\mu_2)^2} \frac{n}{m},\frac{1}{F_{1-\alpha/2,m-1,n-1}} \frac{\sum(x_i-\mu_1)^2}{\sum(y_j-\mu_2)^2} \frac{n}{m}\Big)$

**<font color = red>2. $\mu_1~\&~\mu_2$ are Unknown **

Replace the test statistic by $\frac{S^2_{X}}{S^2_{Y}}$
