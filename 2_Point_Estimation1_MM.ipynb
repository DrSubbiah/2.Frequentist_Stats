## <font color=darkgreen> Method of Moments:

Let X be a random variable ~ $f(X~|~\theta)$ where 
  
$\boldsymbol\theta=(\theta_1, \theta_2........\theta_k)$ where K is the number of parameters
  
Let $\mu^{'}_r = E[X^{r}]$

$\mu^{'}_r$ will be a function of $\theta$
  
Let $X_1, X_2, .....X_n$ be a random sample from $f(X~|~\theta)$ 
  
Now, $m^{'}_r=\frac{1}{n}\sum X_i^r$
  
Solving "k" equations arising from equating these two moments
  
  $\mu^{'}_1=m^{'}_1$
  
  $\mu^{'}_2=m^{'}_2..................$
  
  $\mu^{'}_k=m^{'}_k$
  
Therefore estimators are obtained by comparing population moments and sample moments
  
## <font color=darkgreen>Examples of MOM:

  
### <font color=darkblue> Example 1. 

Let $X_1, X_2, .....X_n$ be a random sample from $N(\mu,\sigma^2)$</span>
  
  $\theta_1 = \mu$
  
  $\theta_2=\sigma^2$
  
  $\mu^{'}_1=E(X)=\mu$
  
  $\mu^{'}_2=E(X^2)=\sigma^2+\mu^2$
  
  $\Rightarrow\frac{\sum X_i}{n}=\mu$
  
  $\frac{\sum{X_i}^2}{n}=\sigma^2+\mu^2$
  
  $\hat\mu_{MM}=\bar X$
  
  $\sigma^2=\frac{\sum{X_i}^2}{n}-\bar X^2$
  
  $\Rightarrow \widehat{\sigma^2}_{MM} = \frac{\sum(X_i-\bar X)^2}{n}$
                                                          
                                                          
### <font color=darkblue>Example 2. 

Let $X_1, X_2, .....X_n$ be a random sample from $f(x~|~\theta)=\theta e^{-\theta x}~~~~~~ x~>~0$</span>

$E(X)=\frac{1}{\theta}$

$\mu^{'}_1=\frac{\sum X_i}{n}=\bar X$

$\Rightarrow\frac{1}{\theta}=\bar X$

$\hat\theta_{MM}=\frac{1}{\bar X}$
 
### <font color=darkblue> Example 3. 

Let $X_1, X_2, ......X_n \sim Uniform(a,b)$

$\mu^{'}_1=\frac{a+b}{2}$</span>
$\mu^{'}_2=\sigma^2+\mu^2$
$\mu^{'}_2 = \frac{{(b-a)}^2}{12}+\mu^2$
$m^{'}_1=\bar X=\mu$
$\frac{a+b}{2}=\bar X$

$\Rightarrow a+b=2\bar X~~~~~~~~~~(1)$

$\mu^{'}_2=\frac{{(b-a)}^2}{12}+\bar X^2$

$\frac{{(b-a)}^2}{12}=\mu^{'}_2-\bar X^2$

$=\frac{1}{n} \sum X_i^2-(\bar X)^2$

$\Rightarrow b-a =2\sqrt 3 \sqrt{\frac{1}{n}\sum X_i^2-(\bar X)^2}$

$=2\sqrt{3}\sqrt{\frac{1}{n}\sum(X_i-\bar X)^2}~~~~~~~~~~(2)$

$(1)~\&~(2) ~ \Rightarrow$

$2b=2[\bar X+\sqrt3 \sqrt{\frac{1}{n}\sum (X_i-\bar X)^2}]$

$b=\bar X+\sqrt3 \sqrt{\frac{1}{n}\sum (X_i-\bar X)^2}$

$a=\bar X-\sqrt3 \sqrt{\frac{1}{n}\sum (X_i-\bar X)^2}]$
                                         
### <font color=darkblue> Example 4. 

Let $X_1, X_2, ......X_n \sim Binomial(m,\theta)$  

$\mu^{'}_1=E(X)=m\theta$

$m^{'}_1=\bar X$

$\mu^{'}_2 = E(X^2) = m\theta(1-\theta) + m^2\theta^2$

$m^{'}_2=\frac{\sum X_i^2}{n}$

$\Rightarrow\bar X=m\theta$

$\hat\theta_{MM} =\frac{\bar X}{m}$

*Now let us estmate m so that $\hat\theta_{MM}$ will be completely obtained* 

$\frac{\sum X_i^2}{m}=m\theta(1-\theta)+m^2\theta^2$

$=m\frac{\bar X}{m}(1-\frac{\bar X}{m})+m^2\frac{(\bar X)^2}{m^2}$

$=\bar X(1-\frac{\bar X}{m}) + \bar X^2$

$=\bar X - \frac{(\bar X)^2}{m}+ \bar X^2$

$\frac{\sum{X_i}^2}{m}-\bar X^2=\bar X-\frac{\bar X^2}{m}$

$\frac{\bar X^2}{m}=\bar X - \frac{1}{n}\sum(X_i-\bar X)^2$

$\Rightarrow \hat m_{MM} =\frac{\bar X^2}{\bar X - \frac{1}{n}\sum(X_i-\bar X)^2}$
                        
                        
### <font color=darkblue> Example 5. 

$X_1......X_n \sim Beta(\alpha,~\beta)$

${M_1}^2=\mu^{'}_1=\frac{\alpha}{\alpha+\beta}$

${M_2}^2=\mu^{'}_2=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}+\Big[\frac{\alpha}{\alpha+\beta}\Big]^2$

$=\frac{\alpha}{\alpha+\beta}~\frac{\beta}{\alpha+\beta}~\frac{1}{\alpha+\beta+1}+{M_1}^2$

$\alpha+\beta=\frac{\alpha}{M_1}$

$\alpha+\beta+1=\frac{\alpha+M_1}{M_1}$

$\frac{1}{\alpha+\beta+1}=\frac{M_1}{\alpha+M_1}]$

$\Rightarrow M_2=M_1(1-M_1)\frac{M_1}{\alpha+M_1}+M_1^2$

$M_2-M_1^2=\frac{M_1^2(1-M_1)}{\alpha+M_1}$

$\alpha+M_1 =\frac{M_1^2(1-M_1)}{M_2-M_1^2}$

$\hat\alpha_{MM} =\frac{M_1^2(1-M_1)}{M_2-M_1^2}-M_1$

$\hat\beta_{MM}=\frac{\hat\alpha_{MM}}{M_1}-\hat\alpha_{MM}$


### <font color=darkblue>Example 6.

$X_1......X_n\sim$ Gamma($\alpha$,$\beta$)

Then $M_1= E(X)=\alpha\beta$ and using $V(X)=\alpha\beta^2$ we get 

$M_2=E(X^2)=\alpha\beta^2+\alpha^2\beta^2$ 

$=M_1\beta+M_1^2$

$\Rightarrow\hat\beta_{MM}=\frac{M_2-M_1^2}{M_1}$

$\Rightarrow\hat\alpha_{MM}=M_1\frac{1}{\hat\beta_{MM}}$
